```python
import os
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook(hide_banner=True)
from bokeh.models import Span
```

```python
basedir = './DOS_GroundState/Testing'

dos_dft = []
first = True
for files in os.listdir(basedir):
    if not files.startswith("."):
        file_path = os.path.join(basedir, files)
        if os.path.isfile(file_path):
            #print(file_path)
            # Need to get header
            dos_data = pd.read_csv(file_path, sep='\s+', skiprows=2) # Gets header
            # print(dos_data.columns.tolist()[1:])
            dos_data = pd.read_csv(file_path, sep='\s+', skiprows=3,names=dos_data.columns.tolist()[1:]) # Removes the # in header
            #print(dos_data.columns.tolist())
            #print(dos_data.head()) 
            if first:
                dos_dft.append(dos_data)
            else:
                dos_data.drop('ENERGY', axis=1,inplace=True)
                dos_dft.append(dos_data)
            first = False

merged_df = pd.concat(dos_dft, axis=1)
print(merged_df)
```
